```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Cost function definition
def compute_cost(X, y, theta_0, theta_1):
    m = len(y)
    h = theta_0 + theta_1 * X  # Predictions
    cost = (1 / (2 * m)) * np.sum((h - y) ** 2)  # Mean Squared Error
    return cost

# Gradient Descent with Cost Tracking
def gradient_descent(X, y, alpha=0.1, iterations=1000):
    m = len(y)
    theta_0, theta_1 = 0, 0  # Initialize parameters
    cost_history = []  # To store cost at each iteration

    for _ in range(iterations):
        h = theta_0 + theta_1 * X  # Predictions
        gradient_0 = (1/m) * np.sum(h - y)  # Gradient for theta_0
        gradient_1 = (1/m) * np.sum((h - y) * X)  # Gradient for theta_1

        # Update parameters
        theta_0 -= alpha * gradient_0
        theta_1 -= alpha * gradient_1

        # Compute and store cost
        cost = compute_cost(X, y, theta_0, theta_1)
        cost_history.append(cost)

    return theta_0, theta_1, cost_history

# Train model and track cost
theta_0, theta_1, cost_history = gradient_descent(X, y)

print(f"Learned parameters: theta_0 = {theta_0:.2f}, theta_1 = {theta_1:.2f}")

# Plot cost function over iterations
plt.plot(range(len(cost_history)), cost_history, label="Cost Function J(θ)")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Convergence")
plt.legend()
plt.show()

# Plot data and regression line
plt.scatter(X, y, label="Data")
plt.plot(X, theta_0 + theta_1 * X, color="red", label="Regression Line")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
